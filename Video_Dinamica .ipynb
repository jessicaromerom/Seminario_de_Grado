VIDEO DINAMICA SIR

```python
import numpy as np
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import pandas as pd

# Leer datos desde el archivo CSV
data = pd.read_csv("posiciones_particulas.csv")

data_valores_ejeX = data['X'].str.strip('[]').str.split(', ', expand=True)
data_valores_ejeY = data['Y'].str.strip('[]').str.split(', ', expand=True)
data_colores = data['Colores'].str.strip('[]').str.replace('"', '').str.split(', ', expand=True)

data_valores_ejeX.to_csv('data_valores_ejeX.csv', index=False)
data_valores_ejeY.to_csv('data_valores_ejeY.csv', index=False)
data_colores.to_csv('data_colores.csv', index=False)

# Cargar los datos desde los archivos CSV
x_n = pd.read_csv("data_valores_ejeX.csv")
y_n = pd.read_csv("data_valores_ejeY.csv")
colores = pd.read_csv("data_colores.csv")

x_n = x_n.transpose()
y_n = y_n.transpose()
colores = colores.transpose()

particles = x_n.shape[0]
time = x_n.shape[1]

fig, ax = plt.subplots(figsize=(1920/100, 1080/100), dpi=100, tight_layout=True)
scatter = ax.scatter(x_n.iloc[:, 0], y_n.iloc[:, 0], marker="o", s=200, alpha=0.6, linewidths=0.5, edgecolors="White", c=colores.iloc[:, 0])

def update_plot(i):
    ax.clear()
    scatter = ax.scatter(x_n.iloc[:, i], y_n.iloc[:, i], marker="o", s=200, alpha=0.6, linewidths=0.5, edgecolors="White", c=colores.iloc[:, i])
    ax.tick_params(labelsize=15)
    ax.set_xticks(np.arange(-50, 51, 50))
    ax.set_yticks(np.arange(-50, 51, 50))
    ax.set_xlim([-50, 50])
    ax.set_ylim([-50, 50])
    ax.set_aspect("equal")

writer = 'ffmpeg-60'
video = animation.FuncAnimation(fig, update_plot, frames=time, interval=70)
video.save('./ video_dinamica_SIR.gif', writer=writer)
```

GRAFICO CURVAS DE LA DINAMICA  SIR

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Dinamica_SIR.csv")

plt.plot(data['S'], label="Susceptibles", linewidth=2, color='blue')
plt.plot(data['I'], label="Infectados", linewidth=2, color='red')
plt.plot(data['R'], label="Recuperados", linewidth=2, color='green')

plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.legend()

plt.show()
```
